```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look cleaner
sns.set()
```

```python
df = pd.read_csv("StudentPerformanceFactors.csv")
df.head(7)
```

```python
df.describe(include='all')
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Hours_Studied'], bins=25)
plt.title("Distribution of Hours Studied")
plt.xlabel("Hours Studied")
plt.ylabel("Frequency")
plt.show()
```

```python
parental_education_count = df['Parental_Education_Level'].value_counts()
plt.figure(figsize=(8,5))
parental_education_count.plot(kind='bar')
plt.title("Number of Students by Parental Education Level")
plt.ylabel("Number of Students", fontsize=12)
plt.xlabel("Parental Education Level", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Exam_Score', y='Attendance', data=df)
plt.title("Relationship between Exam Score and Attendance")
plt.xlabel("Exam Score")
plt.ylabel("Attendance")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[['Exam_Score', 'Sleep_Hours']].corr(), annot=True)
plt.title("Correlation Matrix Between Sleep and Exam Score")
plt.show()
```
